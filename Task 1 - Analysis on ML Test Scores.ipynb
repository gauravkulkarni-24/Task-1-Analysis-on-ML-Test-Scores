### Import Required Libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the Dataset:

```python
df = pd.read_csv(r"scores_data.csv")
df.head()
```

### Extract numeric score:

```python
df['Score_Num'] = df['Score'].str.split('/').str[0].astype(int)
```

```python
df.head()
```

### Replacing Column names:

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

### Dataset Info:

```python
df.info()
```

### Statistical Summary:

```python
df.describe()
```

### Checking Unique Batch:

```python
df['Batch'].unique()
```

- Unique Batches are:
    1. AI_ELITE_7
    2. AI_ELITE_6
    3. AI_ELITE_4

### No. of students in each Batch:

```python
df['Batch'].value_counts()
```

#### Insight:
- This shows how many students are present in each batch.

## Batch Wise Analysis:-

### 1. Average Score per Batch:

```python
df.groupby('Batch')[['Score_Num']].mean()
```

#### Insight:
- Batch **AI_ELITE_7** has the **highest average score**, indicating stronger overall performance.

### 2. Score Distribution per batch:

```python
sns.boxplot(x='Batch', y='Score_Num', data=df)
plt.title("Score Distribution by Batch")
plt.show()
```

#### Insight:
- AI_ELITE_7 shows consistently strong performance with higher median scores and lower variability, indicating better overall understanding and more uniform performance among students.

### 3. Minimum and Maximum score per batch:

```python
df.groupby('Batch')['Score_Num'].agg(['min','max'])
```

#### Insight:
- This shows the performance of batch.

### 4. High Performance (Score >= 6):

```python
df[df['Score_Num']>=6]
```

#### Insight:
- This shows top performing students.

### 5. Low Performance (Score <=3):

```python
df[df['Score_Num']<=3]
```

#### Insight:
- This shows low performing students.

### 6. 

```python
sns.histplot(data=df, x='Score', hue='Batch', bins=7, kde=True)
plt.show()
```

### 7. Compare Batches:

```python
summary = df.groupby('Batch')['Score_Num'].agg(['mean','median','std','min','max'])
summary
```

#### Insight:
- AI_ELITE_7 is the best performing batch with consistently higher scores and fewer weak performers.
- AI_ELITE_6 shows mixed performance with large variability, indicating unequal learning outcomes among students.
- AI_ELITE_4 has comparatively lower average performance with a few students struggling significantly.

### 8. Final Insights & Conclusion

- AI_ELITE_7 performed the best with the highest and most consistent scores.
- AI_ELITE_6 shows mixed performance, with both high and low scores.
- AI_ELITE_4 has the lowest average score and needs more improvement support.
- Overall, AI_ELITE_7 is the strongest batch, while AI_ELITE_4 may require extra attention to improve performance.
